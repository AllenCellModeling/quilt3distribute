```python
import lkaccess
import pandas as pd
```

```python
lk = lkaccess.LabKey(host="aics")
lk
```

```python
all_data = pd.DataFrame(lk.dataset.get_pipeline_4_production_cells())
all_data.head()
```

```python
all_data.shape
```

```python
mito_data = pd.DataFrame(lk.select_rows_as_list("processing", "MitoticAnnotation")).rename(index=str, columns={"Complete": "CompleteSegmentation"})
mito_data.head()
```

```python
mito_states = pd.DataFrame(lk.select_rows_as_list("processing", "MitoticState")).rename(index=str, columns={"Name": "MitoticStateName"})
mito_states.head()
```

```python
data = all_data[[
    'CellId', 'CellIndex', 'CellLine', 'ChannelNumber405', 'ChannelNumber638', 'ChannelNumberBrightfield', 
    'ChannelNumberStruct', 'MembraneSegmentationReadPath', 'NucleusSegmentationReadPath', 
    'NucMembSegmentationAlgorithm', 'NucMembSegmentationAlgorithmVersion', 'PixelScaleX', 
    'PixelScaleY', 'PixelScaleZ', 'PlateId', 'Protein', 'ProteinDisplayName', 'SourceReadPath',
    'Structure', 'StructureSegmentationReadPath', 'StructureSegmentationAlgorithm', 
    'StructureSegmentationAlgorithmVersion', 'Workflow'
]]
data.shape
```

```python
data = data.merge(mito_data[["CellId", "CompleteSegmentation", "MitoticAnnotationId", "MitoticStateId"]], how='left', on='CellId')
data.shape
```

```python
data.head()
```

```python
data = data.merge(mito_states[["MitoticStateId", "MitoticStateName"]], how="left", on="MitoticStateId")
data.shape
```

```python
data.head()
```

```python
data = data.dropna()
data = data.drop_duplicates()
data.shape
```

```python
short_data = data[:1000]
short_data.shape
```
